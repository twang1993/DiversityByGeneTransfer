```python
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
from scipy.integrate import solve_ivp
```

```python
global mu10,mu20,mu1,mu2,gamma,D,eta,kappa
```

```python
def TwoSpecies(t,y):
    s1=y[0]
    s2=y[1]
    p1=y[2]
    p2=y[3]
    lambda1=mu1/mu10-1
    lambda2=mu2/mu20-1

    dydt=[mu1/s1*(s1+lambda2*p1)*s1*(1-s1-gamma*s2)-D*s1,
        mu2/s2*(s2+lambda1*p2)*s2*(1-gamma*s1-s2)-D*s2,
        mu1*(1+lambda2)*p1*(1-s1-gamma*s2)+eta*(s2+p1)*(s1-p1)-kappa*p1-D*p1,
        mu2*(1+lambda1)*p2*(1-gamma*s1-s2)+eta*(s1+p2)*(s2-p2)-kappa*p2-D*p2]
    return dydt
```

```python
gamma=0.9
kappa=0.005
D=0.2
initial=[0.5,0.5,0,0]
timespan=np.arange(0,200.1,0.1)

num=500
mu1s=0.3+0.4*np.random.rand(num)
mu2s=0.3+0.4*np.random.rand(num)
mms=[0.5,0.4,0.3]

etas=np.arange(-14,2,2)
S1=np.zeros((len(mms),len(etas),num))
S2=np.zeros((len(mms),len(etas),num))

for i in range(len(mms)):
    print(i)
    mm=mms[i]
    mu10=mm
    mu20=mm
    for j in range(len(etas)):
        eta=10**int(etas[j])
        for k in range(num):
            mu1=mu1s[k]
            mu2=mu2s[k]
            sol=solve_ivp(TwoSpecies,[0,max(timespan)],initial,method='RK45',t_eval=timespan)
            y=np.transpose(sol.y)
            S1[i,j,k]=y[-1,0]
            S2[i,j,k]=y[-1,1]
```

```python
thresh=0.01
prob=np.zeros((len(mms),len(etas)))
for i in range(len(mms)):
    for j in range(len(etas)):
        pin=0
        for k in range(num):
            if min(S1[i,j,k],S2[i,j,k])>thresh:
                pin=pin+1
        prob[i,j]=pin/num

plt.figure(figsize=(3,3))
for i in range(len(mms)):
    if i==0:
        plt.plot(etas,prob[i,:],'o-',markersize=7,fillstyle='none',color=(0.3467,0.5360,0.6907))
    if i==1:
        plt.plot(etas,prob[i,:],'o-',markersize=7,fillstyle='none',color=(0.9153,0.2816,0.2878))
    if i==2:
        plt.plot(etas,prob[i,:],'o-',markersize=7,fillstyle='none',color=(0.4416,0.7490,0.4322))
plt.box('off')
plt.xlabel('log10(Eta)',fontsize=20)
plt.ylabel('Fold change of coexistence feasibility',fontsize=20)
plt.tick_params(direction='in',top='true',right='true')
plt.savefig('ChangingEta.pdf')
plt.show()
```
